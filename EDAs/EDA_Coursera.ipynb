Importamos librerias necesarias

```python
import pandas as pd
```

cargamos y mostramos nuestro df

```python
df_coursera_reviews=pd.read_csv('D:Coursera_reviews.csv')
df_coursera_reviews
```

Analisis rapido con info y describe

Calculamos los promedios del rating por curso

```python
# Calcular el promedio del rating agrupado por course_id
df_promedio_rating = df_coursera_reviews.groupby('course_id')['rating'].mean().reset_index()

# Renombrar la columna de promedio del rating
df_promedio_rating = df_promedio_rating.rename(columns={'rating': 'promedio_rating'})
df_promedio_rating
```

Cargamos nuestro otro df a partir del csv con la informacion de los cursos y mostramos

```python
df_coursera_courses=pd.read_csv('D:\SoyHenry\Proyecto Integrador 2\Moocs\Coursera_courses.csv')
df_coursera_courses
```

Eliminamos de nuestro df informacion que no vamos a utilizar para nuestro analisis las url

```python
df_coursera_courses.drop(columns='course_url', axis=1, inplace=True)
```

```python
df_coursera_courses
```

unimos ambos dataframes el que tiene la info de los cursos y el que tiene la de los reviews una vez ya calculado el promedio

```python
df_coursera = pd.merge(df_coursera_courses, df_promedio_rating, on='course_id')
```

ordenamos por rating y mostramos nuestro dataset

```python
df_coursera.sort_values('promedio_rating', ascending=False, inplace=True)
df_coursera
```
